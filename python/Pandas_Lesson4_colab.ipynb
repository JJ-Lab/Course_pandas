Convertimos a formato "Fecha":  datetime64[ns]

```python
# from numpy import datetime64
from datetime import datetime
df5['fecha_muestra'] = pd.to_datetime(df5['fecha_muestra'])
```

```python
df5.dtypes
```

```python
df5
```

Me ha rellenado la columna de datetime con "NaT": Not A Time

```python
# Vamos a convertir un string en "datetime"
fecha0='27/07/2022'
```

```python
#  El formato debe coincidir con los datos que le metemos, p.e., %d-%m-%y o %y/%m/%d (%y, p.e.,22)si el año es completo es %Y (2022)
date0 = datetime.strptime(fecha0,'%d/%m/%Y')
print(date0)
print(type(date0))
```

```python
for i in range(df5['id_muestra'].count()):
  id = df5.iloc[i,1]
  serie = df_HRC.loc[df_HRC['CODIGO MUESTRA']==id, 'FECHA MUESTRA']
  print(serie.values)
  fecha = serie.values[0]
  print(id, fecha)
  df5.iloc[i,3]= fecha
  print(id,df5.iloc[i,3])
```
